```python
def decompose_filename(df, column_name='filename'):
    # Create a regular expression pattern to match the parts of the filename
    # This pattern is based on the positions of the parts as specified
    pattern = r'(?P<date>\d{4})vr(?P<vr>\d)_(?P<trials>\d)_(?P<size>\d{3})_a(?P<age>\d{2})_processed\.csv'
    
    # Use the 'str.extract' method to create new columns based on the pattern
    df[['date', 'vr', 'trials', 'size', 'age']] = df[column_name].str.extract(pattern)
    
    return df
```

```python
def add_streak_check_column(df):
    # Check if all three conditions are true for each row and assign 1 if true, else 0
    # df['streak_check'] = ((df['following'] & df['proximity'] & df['rf_behind_vf'])).astype(int)
    df['streak_check'] = ((df['following']  & df['rf_behind_vf'])).astype(int) # 20240403: removed proximity condition
    
    return df

# Example usage
# Assuming 'df' is your DataFrame
df = add_streak_check_column(df)
```
