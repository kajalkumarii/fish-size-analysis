```python
def calculate_streak_counter(df):
    # Initialize the streak_counter column with zeros
    df['streak_counter'] = 0
    counter = 0  # Initialize the counter
    
    for i in range(len(df)):
        if df.loc[i, 'streak_check'] == 1:
            counter += 1  # Increment the counter if streak_check is 1
            df.loc[i, 'streak_counter'] = counter
        else:
            counter = 0  # Reset the counter if streak_check is not 1
    
    return df

# Apply the function to the DataFrame
df = calculate_streak_counter(df)
```

```python
unique_streak_counts = df['streak_counter'].value_counts()
unique_streak_counts_sorted = unique_streak_counts.sort_index()
```

```python
df['stim'] = np.round(df['Stim_Flag'] - 1, 2)
```
